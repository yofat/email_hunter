```python
import pandas as pd

data1  = open("/content/drive/MyDrive/spam email/train.csv","rb")
train = pd.read_csv(data1)

train.head()
```

```python
def process_train_email(email):
    if isinstance(email, str):
        account = email.split('@')[0]
        return account
    return None

all_accounts = []
accounts_with_dot = []
accounts_without_dot = []

for col in train.columns[1::2]:

    train[col] = train[col].apply(process_train_email)

    accounts = train[col].dropna().values
    all_accounts.extend(accounts)

    accounts_with_dot.extend([account for account in accounts if '.' in account])
    accounts_without_dot.extend([account for account in accounts if '.' not in account])

print(f"所有帳號: {all_accounts}")
print(f"包含 '.' 的帳號: {accounts_with_dot}")
print(f"不含 '.' 的帳號: {accounts_without_dot}")
print(f"所有帳號: {len(all_accounts)}")
print(f"包含 '.' 的帳號數量: {len(accounts_with_dot)}")
print(f"不包含 '.' 的帳號數量: {len(accounts_without_dot)}")
```

```python
data2  = open("/content/drive/MyDrive/spam email/prizer.csv","rb")
prize = pd.read_csv(data2)

prize.head()
```

```python
def process_prize_emails(df):
    prize['email'] = prize['email'].str.replace(r'@.*', '', regex=True)

    grouped = prize.groupby('name')['email'].apply(list).to_dict()
    return grouped

prize_grouped = process_prize_emails(prize)

print(prize_grouped)
```

```python
def jaccard_index(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

def find_similar_emails(train_accounts, prize_grouped, threshold=0.5):
    similar_emails = []
    for name, prize_emails in prize_grouped.items():
        for prize_email in prize_emails:
            prize_account = set(prize_email)
            for train_account in train_accounts:
                train_account_set = set(train_account)
                similarity = jaccard_index(train_account_set, prize_account)
                if similarity >= threshold:
                    similar_emails.append((train_account, name, similarity))
    return similar_emails

similar_email_results = find_similar_emails(all_accounts, prize_grouped)

similar_email_df = pd.DataFrame(similar_email_results, columns=['train_account', 'prize_name', 'similarity'])

print("\n相似的電子郵件:")
print(similar_email_df)
```

```python
similarity_stats = similar_email_df['similarity'].describe()
print("相似度的基本統計:")
print(similarity_stats)
```

```python
grouped_analysis = similar_email_df.groupby('prize_name').agg(
    count=('train_account', 'size'),
    average_similarity=('similarity', 'mean')
).reset_index()

print("\n根據 prize_name 的分組分析:")
print(grouped_analysis)
```
